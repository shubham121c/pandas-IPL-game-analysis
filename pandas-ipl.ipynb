```python
mask2=df2['batsman']=='V Kohli'
vk=df2[mask2]
vk.head()
```

```python
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
def run_scored(batsman_name):
    mask2=df2['batsman']==batsman_name
    vk=df2[mask2]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)

    
    
```

```python
run_scored('DA Warner')
```

```python
# to know most agressive batsmen in death over(16-20)
```

```python
mask3=df2['over']>15
death_over=df2[mask3]
```

```python
all_batsman=death_over.groupby('batsman')['ball'].count()
x=all_batsman>200
batsman_list=all_batsman[x].index.tolist()
# to extarct names and convert it into list
```

```python
# runs scored by all 43 batsman
# balls palayed by all 43 batsman
final=df2[df2['batsman'].isin(batsman_list)]
runs=final.groupby('batsman')['batsman_runs'].sum()
balls=final.groupby('batsman')['batsman_runs'].count()

```

```python
sr=runs/balls * 100
sr.sort_values(ascending=False).head(3)
```

```python
# merging two tables
#merge i.e. join tables
```

```python
new=df2.merge(dfx,left_on='match_id', right_on='id')
new.head()
```

```python
# group by using 2 or more columns
aa= new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index()
aa.drop_duplicates(subset='season',keep='first')
```

```python
mask=df2['batsman_runs']==6
six=df2[mask]
six.shape
```
